#### Array - create an array , parameters -> np.array(object, dtype=None, copy=True, order='K', subok=False, ndmin=0)¶

```python
x = np.array([[2,3,1,0], [1,4,6,7], [10,5,3,4]])
print(x)
```

#### Zeros - return a new array of given shape and type , filled with zeros , parameters -> n[.zeros(shape, dtype=float, order='C')

```python
arr = np.zeros((2,5))
print(arr)
```

#### Zeros_like - return an array of zeros with the same shape and type as a given array. parameters -> np.zeros_like(a, dtype=None, order='K', subok=True


```python
np.zeros_like(x)
```

#### Ones - return a new array of given shape and type, filled with ones. parameters -> numpy.ones(shape, dtype=None, order='C')


```python
arr2 = np.ones((4,8))
print(arr2)
```

#### Ones_like - return an array of ones with the same shape and type as a given array. parameters -> np.ones_like(a, dtype=None, order='K', subok=True)


```python
np.ones_like(x)
```

#### Empty - return a new array of given shape and type, without initializing entries. parameters -> np.empty(shape, dtype=float, order='C')



```python
a = np.empty([2, 4], dtype=int)
print(a)
```

#### Empty_like - return a new array with the same shape and type as a given array. parameters -> np.empty_like(prototype, dtype=None, order='K', subok=True)

```python
np.empty_like(a)
```

#### Arange - return evenly spaced values within a given interval. parameters -> np.arange([start, ]stop, [step, ]dtype=None)

```python
np.arange(6,16,2)
```

#### Linspace - return evenly spaced numbers over a specified interval. Returns num evenly spaced samples, calculated over the interval [start, stop]. The endpoint of the interval can optionally be excluded. parameters ->  np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)

```python
np.linspace(2.0, 3.0, num=10, endpoint=False)
```

#### Numpy.random.rand - create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1). parameters -> np.random.rand(d0, d1, ..., dn)

```python
np.random.rand(2,3)
```

#### Numpy.random.randn - return a sample (or samples) from the “standard normal” distribution. parameters -> np.random.randn(d0, d1, ..., dn)

```python
np.random.randn(2, 4) 
```

#### Fromfunction - construct an array by executing a function over each coordinate. The resulting array therefore has a value fn(x, y, z) at coordinate (x, y, z). paramters 0> np.fromfunction(function, shape, **kwargs)

```python
np.fromfunction(lambda i, j : i + j , (4, 4), dtype=int)
```
