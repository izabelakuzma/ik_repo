#### All - test whether all array elements along a given axis evaluate to True. parameters -> numpy.all(a, axis=None, out=None, keepdims=<class 'numpy._globals._NoValue'>)





````python
import numpy as np
arr = np.array([[6, 9], [1, 5], [3, 8]])
print(np.all(arr>5, axis=1))
````

#### Any - test whether any array element along a given axis evaluates to True. parameters -> numpy.any(a, axis=None, out=None, keepdims=<class 'numpy._globals._NoValue'>)

````python
arr = np.array([[6, 9], [1, 5], [3, 8]])
print(np.any(arr>5, axis=1))
````

#### Diff - calculate the n-th order discrete difference along given axis. The first order difference is given by out[n] = a[n+1] - a[n] along the given axis, higher order differences are calculated by using diff recursively. parameters -> numpy.diff(a, n=1, axis=-1)[source]

````python
x = np.array([1, 2, 4, 7, 0])
np.diff(x)
````

#### Cov - estimate a covariance matrix, given data and weights. parameters -> numpy.cov(m, y=None, rowvar=True, bias=False, ddof=None, fweights=None, aweights=None)

````python
a = np.array([ [0.1, 0.3, 0.4, 0.8, 0.9],
               [3.2, 2.4, 2.4, 0.1, 5.5],
               [10., 8.2, 4.3, 2.6, 0.9]
             ])

print( np.cov(a) )
````

#### Round - round an array to the given number of decimals. parameters -> numpy.round_(a, decimals=0, out=None)

````python
a = np.random.random(10)
print(a)
np.round(a,2)
````

Load file baseball.csv from project (see 9.Appendix)

````python
import csv

baseball = []

with open('baseball.csv', newline='') as File:  # with - you have two related operations which you’d like to execute as a pair, with a block of code in between
    reader = csv.reader(File)
    for row in reader:
        baseball.append(row)
        
baseball
````

Import NumPy

````python
import numpy as np
````

Create np_height from np.array which you created from file

````python
np_baseball = np.array(baseball)
np_height = np_baseball[1:, 4].astype(float)
# np_height = np_baseball[:, -4]
np_height

````

Print the mean of np_height

````python


import matplotlib.pyplot as plt

plt.hist(np_height, 200)
plt.hist(np_height.mean())
````

Print the median of np_height

````python
plt.hist(np.median(np_height))
````
