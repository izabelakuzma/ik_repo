Print cities with population > 10 000


````python
import pandas as pd
df = pd.read_csv('lesserpoland.csv')
df = pd.read_csv('lesserpoland.csv', names=['City','Counts','NaN'])
to_drop = ['NaN']
df.drop(to_drop, inplace=True, axis=1)
df.dropna(subset=['Counts'], inplace=True)
df_filtered = df[df['Counts'] > 10000]
print(df_filtered)
````
